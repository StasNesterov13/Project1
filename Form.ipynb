```python
import pandas as pd
import datetime
import numpy as np
```

```python
borename = str(input())
```

```python
df = pd.read_excel(f'Z:2-UGIR\ONtrim\Нестеров С\Куюмбинское месторождение\\{borename}.xlsx')
```

```python
df = df[df['Парам/Дата'].isin(['Состояние','Qж, м3/сут', 'Обв, %', 'Qн, т/сут', 'Qгаз, м3/сут', 'ГФ, м3/т', 'Рприем, атм'])]
df = df.iloc[:, 5:]
df = df.rename(columns={'Парам/Дата':'Дата'})
df = df.assign(idx=df.groupby('Дата').cumcount()).set_index('idx')
```

```python
df
```

```python
res = pd.concat([df.loc[i].set_index('Дата').dropna(how='all', axis='columns') for i in list(range(43, -1, -1))], axis=1)
res = res.T
res.index = pd.date_range(start='01/01/2020', end='08/31/2023').date.tolist()
res.columns.name = None
res.index.name = 'Дата'
```
